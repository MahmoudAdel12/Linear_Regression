# Check OLS assumtions

https://towardsdatascience.com/verifying-the-assumptions-of-linear-regression-in-python-and-r-f4cd2907d4c0

1_ Must be linear relation  between dependent variable and independent variable  





2_ No Endogeneity ( Means their is no relation between independent variables and and error )





3_ Normality and Homoscedasticity (equal variance) of residuals

     ~ N(0 , ^2 )
    
* Noramlity 

* Zero mean

* Homoscedasticity
    


    When residuals do not have constant variance (they exhibit heteroscedasticity)
    
    We can also use two statistical tests: Breusch-Pagan and Goldfeld-Quandt. In both of them, the null hypothesis assumes 
    homoscedasticity and a p-value below a certain level (like 0.05) indicates we should reject the null in favor of 
    heteroscedasticity.
    
   
   
   
4_ No autocorrelation of residuals (If the data doesn't comed from series data then there is no autocorrelation )
    
    Some notes on the Durbin-Watson test:

        the test statistic always has a value between 0 and 4
        value of 2 means that there is no autocorrelation in the sample
        values < 2 indicate positive autocorrelation, values > 2 negative one.
        
        
5_No Multicollinearity

    In other words, the features should be linearly independent. What does that mean in practice? We should not be able to use a 
    linear model to accurately predict one feature using another one. Lets take X1 and X2 as examples of features. It could 
    happen that X1 = 2 + 3 * X2

# Check multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 

variables = data_cleaned[['Mileage','Year' ,'EngineV']]
vif = pd.DataFrame()

vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]

vif["features"] = variables.columns
```

```python
vif

# VIF  [1, +inf]

# VIF = 1: no multicollinearity

# 1 < VIF < 5 : perfectly okay

# VIF > 10 : is unacceptable

```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis = 1)
```

# Create dummy variables

    If we have n category we should create n-1 dummies (So in this case we will use drop_first to make sure that Audi will't 
    take dummy )
    
    If we include a sperate dummy for each category we will intreduce multicollinearity to the regression !!!

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies.head() 
```

```python
variables = data_with_dummies[['Brand_BMW', 'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas', 'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes',]]
vif = pd.DataFrame()

vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]

vif["features"] = variables.columns
```

```python
cols = ['Log Price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
```

```python
data_preprocessed.head()
```

# Linear regression model

# Declear the input and target

```python
target = data_preprocessed['Log Price']
continus_inputs =  data_preprocessed[['Mileage','EngineV']]
```

# Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit(continus_inputs)
```

```python
inputs_scaled = scaler.transform(continus_inputs)
inputs_scaled
```

```python
continus_inputs.columns
```

```python
inputs_scaled_DataFrame = pd.DataFrame(inputs_scaled,columns = continus_inputs.columns)
inputs_scaled_DataFrame
```

```python
categorical_data = data_preprocessed.drop(['Log Price','Mileage','EngineV'], axis = 1)
categorical_data
```

```python
inputs = pd.concat([inputs_scaled_DataFrame,categorical_data],axis=1 )
```

```python
inputs
```

# Train test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =  train_test_split(inputs, target, test_size=0.2, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)

plt.xlabel('Taregets (y_train)', size = 18)

plt.ylabel('Predictions (y_hat)', size = 18)

plt.xlim(6,13)

plt.ylim(6,13)
```

# Residual PDF

```python
reg.score(x_train, y_train)
```

* Positive weight shows that feature increnses the target 



* Negative weight shows that feature decrenses the target




2_ Dummy variables (Dummies not compared to continuous variable it compared to benchmark (droped value in categories in our example is audi ) ) : 

* A positive weight shows that respecive category (Brand) is more expensive than the benchmark (Audi)



* A positive weight shows that respecive category (Brand) is less expensive than the benchmark (Audi)
